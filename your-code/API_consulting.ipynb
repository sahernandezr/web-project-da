For the API portion of the project will need to make calls to your chosen API, successfully obtain a response, request data, convert it into a Pandas data frame, and export it as a CSV file.

```python
import requests
from bs4 import BeautifulSoup
import re
import time
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#Chosen API: Harry Potter API https://hp-api.herokuapp.com/
#endpoint: Hogwarts staff during the book series
url = 'https://hp-api.herokuapp.com/api/characters/staff'

response = requests.get(url)
results = response.json()

results0 = results[0]
results0
```

```python
results0.keys()
columns = [key for key in results0.keys()]
columns
```

```python
df_hogwarts_staff = pd.DataFrame(results, columns=[i for i in columns])
df_hogwarts_staff
```

```python
df_hogwarts_staff.to_csv('output/API-HP-hogwarts-staff.csv')
```
