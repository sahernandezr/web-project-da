For the web scraping portion of the project, you will need to scrape the HTML from your chosen page, parse the HTML to extract the necessary information, and either save the results to a text (txt) file if it is text or into a CSV file if it is tabular data.

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://disenosparabordar.com/hilos/para-bordar/mouline-special-117/?filter_gama='
```

```python
info = requests.get(url).content
soup = BeautifulSoup(info, "html.parser")
soup
```

```python
products = soup.find_all('div', {'class': 'product-card'})
products
```

```python
images = [img for img in soup.select('div.img')]
images
```

```python
images_links = []
for i in range(0,len(images)):
    image = re.sub('"/></div>.*?','',re.sub('^.*?src="','',re.sub('\\n','',str(images[i]))))
    images_links.append(image)
    i+=1
images_links
```

```python
links = [a for a in soup.select('div.title')]
links
```

```python
product_links = []
for i in range(2, len(links)): #0 and 1 are cart and header, not products
    link = str(links[i])
    link = re.sub('\n','',link)
    link = re.sub('.*href="','',link)
    link = re.sub('/">\r.*','',link)
    product_links.append(link)
    i=+1
product_links
```

```python
names = [a.text for a in soup.select('div.title')]
names
```

```python
names_list=[]
for i in range(2,len(names)): #0 is Canasta de compras and 1 is name of page: "Mouliné Special 117"
    name = str(names[i])
    name = re.sub("\t|\n|\r","",name)
    name = name.rstrip()
    names_list.append(name)
    i=+1
names_list
```

```python
import pandas as pd
dict = {'name': names_list, 'image_link': images_links, 'product_link': product_links} 

df_hilos_dmc = pd.DataFrame(dict)
df_hilos_dmc
```

```python
df_hilos_dmc.to_csv('output/webscrapping_hilos_dmc_100.csv')
```
